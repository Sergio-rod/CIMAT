```python
import numpy as np
```

```python
list_n = [9,4,3,2,3,2,41,1,24,42,4]
list_n
```

```python
uniques = set(list_n)
```

```python
dx_counts = {}
for unique in uniques:
    dx_counts[unique] = 0
```

```python
for n in list_n:
    dx_counts[n] += 1
    
```

```python
sorted_list = []
for idx,count in dx_counts.items():
    aux_list = [idx]
    min_number = idx
    if idx<min_number:
        min_number = idx
    aux_list = aux_list*count
    sorted_list += sorted_list
    
```
